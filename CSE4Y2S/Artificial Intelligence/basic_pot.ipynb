```python
import numpy as np
import matplotlib.pyplot as plt
```

define functions

y = ax + b


```python
def func1(x,a,b):
    return a*x+b
```

y = a*sin(x) + b

```python
def func2(x,a,b):
    # return a*np.sin(x*(np.pi/180))+b
    return a*np.sin(x)+b
```

y = a*e^x + b

```python
def func3(x,a,b):
    return a*np.exp(x)+b
```

y = ax^2 + b

```python
def func4(x,a,b):
    return a*(x**2)+b
```

main function


```python
a,b=10,5
step=0.1
x=np.arange(-10,10,step)
# print(len(x)) 200 points

y1=func1(x,a,b)
y2=func2(x,a,b)
y3=func3(x,a,b)
y4=func4(x,a,b)

# plot the functions
plt.subplot(2,2,1)
plt.plot(x,y1,label='y=ax + b')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,y2,label='y=asin(x) + b')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y3,label='y=ae^x + b')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y4,label='y=ax^2 + b')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()

plt.tight_layout()
```
